Resume matcher with Job Description

```python
import turicreate
```

```python
product=turicreate.SFrame('/content/sample_data/amazon_baby (1).csv')
```

```python
selected_words=['HTML','CSS','Javascript','React JS','MONGO DB','SQL','DJANGO']
```

```python
def rat(dict):
  x=0
  for i in selected_words:
    if i in dict:
      x+=1
  return x
```

```python
product['rating']=product['skills'].apply(lambda x:rat(x))
```

```python
product
```

```python
include_words=['full stack','MERN Stack','webdeveloper','web developer','University','College','graduate','grad','CSE','computer science','B.tech','responsible','proficient','passionate','developing','designing','websites','company','creating','maintaining','maintained','HTML','CSS','Javascript','React JS','MONGODB','SQL','DJANGO',]
```

```python
product['word_count']=turicreate.text_analytics.count_words(product['intro'])
```

```python
def var_cou(dict,word):
  if word not in dict:
    return 0
  return dict[word]
```

```python
for i in include_words:
  product[i]=product['word_count'].apply(lambda x:var_cou(x,i))
```

```python
product
```

```python
product['sentiment']=product['rating']>=len(selected_words)//2
```

```python
include_words=include_words.append('rating')
```

```python
train_data,test_data=product.random_split(.8,seed=0)
```

```python
model=turicreate.logistic_classifier.create(train_data,target='sentiment',features=include_words,validation_set=test_data)
```

```python
product
```

```python
product['percentage of match']=(model.predict(product,output_type='probability'))*100
```

```python
product=product.sort('percentage of match',ascending=False)
```
